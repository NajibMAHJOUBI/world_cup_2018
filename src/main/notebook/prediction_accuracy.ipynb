```python
import os
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
year = 2014
path_prediction = os.path.join("../test/prediction/stacking", str(year))
stages = ["1st_stage", "2nd_stage", "3rd_stage", "4th_stage", "5th_stage", "6th_stage"]
stacking_models = sorted(os.listdir(os.path.join(path_prediction, stages[0])))
stacking_classifiers = ["decision_tree", "random_forest", "logistic_regression"]
```

## Prediction

```python
dic_accuracy = {classifier: [] for classifier in stacking_classifiers}

for classifier in stacking_classifiers[:]:
    print("Classifier: {0}".format(classifier))
    for model in stacking_models[:]:
        list_stages = []
        for stage in stages:
            path = os.path.join(path_prediction, stage, model, classifier)
            list_files = os.listdir(path)
            csv_file = filter(lambda x: x.endswith(".csv"), list_files)[0]
            data = pd.read_csv(os.path.join(path, csv_file), path, header=0, delimiter=";", index_col=False)
            list_stages.append(data)
        data_concat = pd.concat(list_stages)
        label_prediction = map(lambda x: (float(x.split(",")[1]), float(x.split(",")[2])), data_concat.to_dict('list').values()[0])
        label = map(lambda x: x[0], label_prediction)
        prediction = map(lambda x: x[1], label_prediction)
        print("  Model: {0} - Accuracy = {1}".format(model, accuracy_score(label, prediction, normalize=True)))
        dic_accuracy[classifier].append((model, accuracy_score(label, prediction, normalize=True)))
```

```python
for key in dic_accuracy.keys():
    print("Stacking classifier: {0}".format(key))
    max_accuracy = max(map(lambda x: x[1], dic_accuracy[key]))
    print("Max accuracy: {0}".format(max_accuracy))
    models = map(lambda x: x[0], filter(lambda x: x[1] == max_accuracy, dic_accuracy[key]))
    print(models)
    print("")
```
